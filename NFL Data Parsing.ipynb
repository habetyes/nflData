```python
# df.flex_yards.unique()
df["flex_yards"].replace(np.nan, 0, inplace=True)
df["touchdown"].replace(np.nan, 0, inplace=True)
df["men_in_box"].replace(np.nan, 0, inplace=True)
df["players_rushed"].replace(np.nan, 0, inplace=True)
df["blitz"].replace(np.nan, 0, inplace=True)
df["touchdown"].replace(np.nan, 0, inplace=True)
df["QB_attempt"].replace(np.nan, 0, inplace=True)
df["QB_complete"].replace(np.nan, 0, inplace=True)
df["QB_air_yards"].replace(np.nan, 0, inplace=True)
df["QB_yards"].replace(np.nan, 0, inplace=True)
df["QB_yards"].replace(np.nan, 0, inplace=True)
df["QB_firstdown"].replace(np.nan, 0, inplace=True)
df["QB_blitz"].replace(np.nan, 0, inplace=True)
df["QB_hurry"].replace(np.nan, 0, inplace=True)
df["QB_knockdown"].replace(np.nan, 0, inplace=True)
df["QB_pocket_time"].replace(np.nan, 0, inplace=True)
df["QB_inside20"].replace(np.nan, 0, inplace=True)
df["QB_goaltogo"].replace(np.nan, 0, inplace=True)
df["flex_attempt"].replace(np.nan, 0, inplace=True)
df["flex_yards"].replace(np.nan, 0, inplace=True)
df["flex_brokentackle"].replace(np.nan, 0, inplace=True)
df["flex_yacontact"].replace(np.nan, 0, inplace=True)
df["flex_target"].replace(np.nan, 0, inplace=True)
df["flex_reception"].replace(np.nan, 0, inplace=True)
df["flex_yacatch"].replace(np.nan, 0, inplace=True)
df["flex_firstdown"].replace(np.nan, 0, inplace=True)
df["flex_inside20"].replace(np.nan, 0, inplace=True)
df["flex_goaltogo"].replace(np.nan, 0, inplace=True)
df["flex_scramble"].replace(np.nan, 0, inplace=True)
df["flex_kneel"].replace(np.nan, 0, inplace=True)
df["flex_dropped"].replace(np.nan, 0, inplace=True)
df["DB_target"].replace(np.nan, 0, inplace=True)
df["DB_completion"].replace(np.nan, 0, inplace=True)
df["DB_tackle"].replace(np.nan, 0, inplace=True)
```

```python
df["flex_opportunities"] = df["flex_attempt"] + df["flex_target"]
df["flex_touches"] = df["flex_attempt"] + df["flex_reception"] 
```

```python
df.loc[df.plays_in_drive == df.drive_play_number, 'last_play_in_drive'] = True
df.loc[df.plays_in_drive != df.drive_play_number, 'last_play_in_drive'] = False
```

```python
df.loc[df.flex_position == "TE", 'Jake_Points'] = ((df["touchdown"] * 6) + (df["flex_yards"] *.1) + (df["flex_reception"] * 2) + (df["flex_firstdown"]))
df.loc[df.flex_position != "TE", 'Jake_Points'] = ((df["touchdown"] * 6) + (df["flex_yards"] *.1) + (df["flex_reception"] * 1) + (df["flex_firstdown"]))
df["PPR_Points"] = ((df["touchdown"] * 6) + (df["flex_yards"] *.1) + (df["flex_reception"] * 1))
df["Standard_Points"] = ((df["touchdown"] * 6) + (df["flex_yards"] *.1))
```
